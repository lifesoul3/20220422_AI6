```python
import pandas as pd
import seaborn as sns

print(pd.__version__)
print(sns.__version__)
```

```python
#sns를통한 데이터 로드
df=sns.load_dataset('mpg')
df
```

```python
sns.pairplot(data=df)
```

```python
df_sample=df.sample(100,random_state=40)
sns.pairplot(data=df_sample,hue="origin")
```

```python
sns.pairplot(df,y_vars=['mpg'],hue='origin')
```

```python
sns.scatterplot(data=df,x='model_year',y='mpg')
```

```python
sns.lineplot(data=df,x='model_year',y='mpg',ci=100)
```

```python
sns.relplot(data=df,x='model_year',y='mpg',hue='origin',col='origin')
```

```python
sns.relplot(data=df)
```

```python
sns.relplot(data=df,x='weight',y='mpg',hue='origin',col='origin',kind='line')
```

```python
corr=df.corr()
corr
```

```python
sns.heatmap(corr,cmap='autumn',annot=True)
```
