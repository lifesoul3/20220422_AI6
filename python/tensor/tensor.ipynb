```python
import pandas as pd
import seaborn as sns
print(pd.__version__)
print(sns.__version__)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
print(tf.__version__)
```

```python
dataset=sns.load_dataset("mpg")
dataset.shape
```

```python
#결측치
dataset.isnull().sum()
#dataset.isna().sum()
```

```python
#결측치 삭제
dataset=dataset.dropna()
dataset.shape
```

```python
dataset.info()
dataset
```

```python
#행삭제
#name 삭제
dataset.pop("name")
```

```python
dataset
```

```python
#origin 열 범wn형 원-핫 인코딩으로 변환
origin=dataset.pop("origin")
```

```python
dataset['usa']=(origin=='usa')*1.0
dataset['europe']=(origin=='europe')*1.0
dataset['japan']=(origin=='japan')*1.0
```

```python
dataset.sample(20,random_state=42)
```

```python
#데이터 분할
#학습 - 8 - train_dataset
#평가 - 2 - test_tastset
train_dataset=dataset.sample(frac=0.8,random_state=42)
test_dataset=dataset.drop(train_dataset.index)
train_dataset.info()
test_dataset.info()
```

```python
sns.pairplot(train_dataset[['mpg','cylinders','displacement','weight']])
```

```python
#기술 통계
train_stats=train_dataset.describe()
train_stats.pop('mpg')
train_stats=train_stats.transpose()
train_stats
```

```python
# 레이블 - 예측할려는 칼럼
# 특성 - 다른 컬럼
# 특성과 레이블을 분리

train_labels=train_dataset.pop('mpg')
test_labels=test_dataset.pop('mpg')
```

```python
train_dataset.head(10)
train_labels.head(10)
train_stats
```

```python
#데이터 정규화
#연산속도와 예측율 상승
def norm(x):
    return ((x-train_stats['mean'])/train_stats['std'])
normed_train_data=norm(train_dataset)
normed_test_data=norm(test_dataset)
normed_train_data
```

```python
#AI 모델 만들기
#deeprunning 형태로 만들기
def build_model():
    model=keras.Sequential([
        layers.Dense(64,input_shape=[len(train_dataset.keys())],activation='relu'),
        layers.Dense(64,activation='relu'),
        layers.Dense(1)
    ])#레이어 생성
    optimizer=tf.keras.optimizers.RMSprop(0.001)#오답 처리
    model.compile(loss='mse',optimizer=optimizer,metrics=['mae','mse'])#컴파일
    return model
    
```

```python
#모델 확인
model=build_model()
model.summary()
```

```python
#모델을 통한 훈련
#설정 1000번 학습(한번훈련시 epoch라는 용어를 씀),1번의 학습을 할때 점을 하나 찍는다
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self,epoch,logs):
        if epoch%100==0:
            print("")
        print('.',end="")#개행 제거
        
EPOCHS=1000
model=build_model()
history=model.fit(
    normed_train_data,train_labels,epochs=EPOCHS,validation_split=0.2,verbose=0,callbacks=[PrintDot()]
)
```

```python
hist=pd.DataFrame(history.history)
hist['epoch']=history.epoch
hist
```

```python
#학습 1000번에 대한 가성비
#학습을 진행하면서 오차율을 시각화
def sns_hist(history):
    hist=pd.DataFrame(history.history)
    hist['epoch']=history.epoch
    sns.lineplot(data=hist,x='epoch',y='mae')
    sns.lineplot(data=hist,x='epoch',y='mse')
sns_hist(history)
#그래프를 보면 200정도까지만 해도 될것이다
```

```python
#model.fit을 수정하여 검증 점수가 향상되지않으면 자동으로 훈련 종료
#rebuild
model=build_model()
#patience x회 이상부터 monitor실행 
early_stop=keras.callbacks.EarlyStopping(monitor='val_loss',patience=50)
history=model.fit(
    normed_train_data,train_labels,epochs=EPOCHS,validation_split=0.2,verbose=0,callbacks=[PrintDot(),early_stop]
)
```

```python
hist=pd.DataFrame(history.history)
hist['epoch']=history.epoch
hist
```

```python
sns_hist(history)
```

```python
#model성능 평가 
loss,mae,mse=model.evaluate(normed_test_data,test_labels,verbose=2)
print('테스트 세트의 평균 절대 오차:',mae,'MPG')
print('테스트 세트의 평균 절대 오차:{:.4f} MPG'.format(mae))
```
