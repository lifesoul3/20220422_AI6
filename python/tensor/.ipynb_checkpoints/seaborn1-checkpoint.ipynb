```python
#import pandas,seaborn
#check version
import pandas as pd
import seaborn as sns
print(pd.__version__)
print(sns.__version__)
```

```python
#loading data(mpg)
#check info
df=sns.load_dataset('mpg')
df.info()
```

```python
#using histogram
df.hist(figsize=(10,8),bins=100)
```

```python
#using skew kurt agg
#df.pop('origin')
#df.pop('name')
df.skew()
df.kurt()
df['mpg'].agg(['skew','kurt'])
```

```python
#using plot(5)
sns.violinplot(data=df,x='mpg')
```

```python
sns.boxplot(data=df['mpg'])
```

```python
sns.lineplot(data=df,x='mpg',y='horsepower')
```

```python
sns.boxenplot(data=df['mpg'])
```
