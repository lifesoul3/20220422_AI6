```python
import pandas as pd
import seaborn as sns
```

```python
print(pd.__version__)
print(sns.__version__)
```

```python
# df-> pandas -> data frame
df=sns.load_dataset("mpg")
df
```

```python
df.info()
```

```python
# 히스토그램 - 표로 되어 있는 도수 분포
# 함수 사용법->shift+tab
df.hist(figsize=(10,8), bins=100)
```

```python
# 비대칭도(왜도)
# 왜도가 음수일 경우에는 오른쪽으로 기운다
# 왜도가 양수일 경우에는 왼쪽으로 기운다
# 평균과 중앙값이 같으면 0
# skew()-왜도 구하기
df.skew()
```

```python
# 첨도
# 확률 분포의 뾰족한 정도를 나타내는 척도
# 데이터가 어느 곳에 집중적으로 몰려 있는가를 측정할 때 사용
# 첨도값이 3에 가까울 수록 정규 분포에 가깝다
# K<3 더 뾰족한 분포(꼬리가 얇다)
# K>3 더 완만하고 납작한 분포(꼬리가 두꺼운 분포)
# kurt()
df.kurt()
```

```python
# seaborn->히스토그램->displot()
sns.displot(data=df,x="mpg",kde=True)
```

```python
# kdeplot() 선으로 밀도 함수 표현하기
sns.kdeplot(data=df,x="mpg",shade=True)
```

```python
# mpg 값에 대해 skew, kurt 값 구하기
# 하나의 칼럽 객체 -> 시리즈(Series)
df["mpg"].agg(["skew","kurt"])
```

```python
# boxplot으로 mpg의 사분위 수 시각화
sns.boxplot(data=df,x="mpg")
```
