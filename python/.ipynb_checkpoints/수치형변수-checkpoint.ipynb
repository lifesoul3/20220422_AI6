```python
# ë¹„ëŒ€ì¹­ë„(ì™œë„)
# ì™œë„ê°€ ìŒìˆ˜ì¼ ê²½ìš°ì—ëŠ” ì˜¤ë¥¸ìª½ìœ¼ë¡œ ê¸°ìš´ë‹¤
# ì™œë„ê°€ ì–‘ìˆ˜ì¼ ê²½ìš°ì—ëŠ” ì™¼ìª½ìœ¼ë¡œ ê¸°ìš´ë‹¤
# í‰ê· ê³¼ ì¤‘ì•™ê°’ì´ ê°™ìœ¼ë©´ 0
# skew()-ì™œë„ êµ¬í•˜ê¸°
df.skew()
```

```python
# ì²¨ë„
# í™•ë¥  ë¶„í¬ì˜ ë¾°ì¡±í•œ ì •ë„ë¥¼ ë‚˜íƒ€ë‚´ëŠ” ì²™ë„
# ë°ì´í„°ê°€ ì–´ëŠ ê³³ì— ì§‘ì¤‘ì ìœ¼ë¡œ ëª°ë ¤ ìžˆëŠ”ê°€ë¥¼ ì¸¡ì •í•  ë•Œ ì‚¬ìš©
# ì²¨ë„ê°’ì´ 3ì— ê°€ê¹Œìš¸ ìˆ˜ë¡ ì •ê·œ ë¶„í¬ì— ê°€ê¹ë‹¤
# K<3 ë” ë¾°ì¡±í•œ ë¶„í¬(ê¼¬ë¦¬ê°€ ì–‡ë‹¤)
# K>3 ë” ì™„ë§Œí•˜ê³  ë‚©ìž‘í•œ ë¶„í¬(ê¼¬ë¦¬ê°€ ë‘êº¼ìš´ ë¶„í¬)
# kurt()
df.kurt()
```

```python
# mpg ê°’ì— ëŒ€í•´ skew, kurt ê°’ êµ¬í•˜ê¸°
# í•˜ë‚˜ì˜ ì¹¼ëŸ½ ê°ì²´ -> ì‹œë¦¬ì¦ˆ(Series)
df["mpg"].agg(["skew","kurt"])
```
