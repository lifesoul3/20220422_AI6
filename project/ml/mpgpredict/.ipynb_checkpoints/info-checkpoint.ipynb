```python
def loading():
    data=sns.load_dataset('mpg')
    data.loc[data['horsepower'].isna(),'horsepower']=data['horsepower'].mean()
    data.to_csv(os.path.join('data/','data.csv'),index=False)
    return data

data=loading()
```
